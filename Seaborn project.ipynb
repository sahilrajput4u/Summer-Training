```python
amo['Date'] = pd.to_datetime(amo['Date'])

sns.lineplot(data=amo,x='Date',y='Product line')
plt.title('Product Line Over Time')
plt.xlabel('Date')
plt.ylabel('Products')
plt.show()
```

```python
amo['Date'] = pd.to_datetime(amo['Date'])

sns.relplot(data=amo,x='Date',y='Product line',kind = 'scatter')
plt.title('Product Line Over Time')
plt.xlabel('Date')
plt.ylabel('Products')
plt.show()
```

COVID-19 Data Analysis

```python
co = pd.read_csv('/content/Latest Covid-19 India Status.csv')
co
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(data=co,x='State/UTs',y='Population')
plt.title('Population Over State')
plt.xlabel('State/UTs data')
plt.ylabel('Population data')
plt.show()
```

```python
sns.barplot(data=co,x='Deaths',y='Discharged')
plt.title('Deaths and Discharge ')
plt.xlabel('Deaths')
plt.ylabel('Discharged data')
plt.show()
```

```python
sns.pairplot(data=co)
plt.show()
```

```python
sns.barplot(data=co,x='State/UTs',y='Total Cases')
plt.title('Total Cases Over State')
plt.xlabel('State/UTs data')
plt.ylabel('Total Cases data')
plt.show()
```

Heatmaps

```python
sns.heatmap(co.corr(numeric_only=True))
plt.show()
```
